```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize = (15,7))
x = ["Rahmatshah", "Kusalperea", "Root", "Poodran", "Duplessis", "Babar", "Warner", "Shakib", "Rohit", "Williamson"]
y = [14.8, 18.16, 19.07, 20.01, 21.06, 24.51, 25.02, 28.25, 29.05, 30.23]

plt.bar(x,y)
plt.xlabel("Players")
plt.ylabel("% of team's runs scored by top scorer")
plt.title("The world cup's big guns")
plt.show();
```
